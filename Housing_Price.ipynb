## Plots
The most important plots you'd be using most of the time are 
- Histograms: to get a visual on your data distribution. We'd want to know if they're normally distributed or skewed. Can also be used to detect outliers
- Boxplots: also to understand the data distribution and detect outliers(shows outliers clearly)
- Scatterplots and Heatmaps: helps us to understand the relationship(correlation) between continuous variables


```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#Histogram

```
# Say we want to plot a histogram for all columns in numeric dataframe
numeric.hist(bins=10,figsize=(20,15))
plt.show()
```

```
# histogram for the '2ndFlrSF' column in the numeric dataframe
# the bins parameter is used to set the boundaries in the plot 
# a good practice is to set your number of bins = sqrt(number of entries)
plt.hist(numeric['2ndFlrSF'],bins=10)
plt.style.use('ggplot')
plt.show()
```

From the above plot , we can deduce that this particular column is not normally distributed and is skewed to the left. We can also say that it also contains outliers as there are a very few extreme values towards the 2000 mark. These can be solved using standardization and logloss tarnsfomation techniques and outliers can be solved by either removing or some other techniques.

## Scatterplots, Heatmaps and Correlation

### Scatterplots
This are used for showing the relationship/correlation between two continuous variables. They're very useful in features selection when trying to improve your model especially when your target is a continuous variable. 

[Click here to check image](https://drive.google.com/file/d/1IiqFJ1TsMCrQbRKxQBz0B1Bx57F-m8Z2/view?usp=sharing)

From the image above we can see how to understand scatterplots. 

```
# Plotting GarageArea and MSSubClass against SalePrice
plt.scatter(x=numeric['GarageArea'],y=numeric['SalePrice'])
plt.xlabel('GarageArea')
plt.ylabel('Sale Price')
plt.show()
```

From the above plot, we can se that GarageArea correlates positively with SalePrice(target), therfore we can say that GarageArea is a pretty good predictor of SalePrice

```
plt.scatter(numeric['MSSubClass'], numeric['SalePrice'], color='g')
plt.xlabel('MSSubSclass')
plt.ylabel('Sale Price')
plt.show()
```

From the above, we can see that there's no correlation between MSSubClass and Sale Price , therefore, it is really not a good predictor of our target and might affect our model performance

```
#Scatterplot with seaborn. Parameter 'hue=' allows us to group our visualization by anotjer feature
# Check Seaborn documentation for more details
sns.scatterplot(x='BsmtFinSF1',y='SalePrice', hue=objects['MSZoning'],data=numeric)
```

```
#Seaborn regplot() allows us to draw a regression line in our scatterplot
sns.regplot(x='BsmtFinSF1',y='SalePrice',data=numeric)
```

```
#to plot a scatterplot for all columns
for col in numeric.columns:
  plt.scatter(numeric[col], numeric['SalePrice'], color='g')
  plt.xlabel(col)
  plt.ylabel('Sale Price')  
  plt.show()
```

### Heatmaps
Heatmaps gives a better way of performing the above process and gives a more general sense of our data.

```
corrMatrix=numeric.corr()
print(corrMatrix)
```

The Above is called a correlation matrix. By calling the .corr function, it returns a table of the correlation of each columns against themselves(Pairwise correlation). 
* A correlation of values >=0.5 up till 1 signifies a positive correlation 
* A correlation of values between > -0.5 up till 0 up till < 0.5 signifies no correlation at all and most times such features aren't relevant to our model
* A correlation of values <= -0.5 up till -1 shows a negative correlation

the threshold of -0.5 and 0.5 can be changed however you want but most times, 0.5 is best
Positively and negatively correlated features show relevance of such features to our model. We're interested inthe correlation of our independent features against our target variable.

```
# we pass the above correlation matrix into our heatmap to generate a plot 
plt.subplots(figsize=(20,15))
sns.heatmap(corrMatrix,annot=True)
plt.show()
```

Heatmaps can also be used to show relationship between two or more continuous variables.
